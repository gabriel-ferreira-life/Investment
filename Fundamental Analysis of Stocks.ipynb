```python
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request


SCRIP = ['INFY', 'LTI', 'MSFT', 'NIO']
for i in range(len(SCRIP)):
    
    link = f'https://www.screener.in/company/{SCRIP[i]}'
    hdr = {'User-Agent':'Mozilla/5.0'}
    req = Request(link,headers=hdr)
 
    try:
        page=urlopen(req)
        soup = BeautifulSoup(page)

        div_html = soup.find('div',{'class': 'company-ratios'})
        ul_html = div_html.find('ul',{'id': 'top-ratios'})
        market_cap = 0.0

        for li in ul_html.find_all("li"):
            name_span = li.find('span',{'class':'name'})
            if 'Market Cap' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                market_cap = float(num_span) if (num_span != '') else 0.0

            if 'Current Price' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                Current_Price = float(num_span) if (num_span != '') else 0.0

            if 'High / Low' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')

                num_span1 = li.find('span',{'class':'number'}).find_next('span')
                num_span1 = num_span1.text.replace(',', '')

                High = float(num_span) if (num_span != '') else 0.0
                Low = float(num_span1) if (num_span1 != '') else 0.0

            if 'Stock P/E' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                Stock_PE = float(num_span) if (num_span != '') else 0.0

            if 'Book Value' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                Book_Value = float(num_span) if (num_span != '') else 0.0

            if 'Dividend Yield' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                Dividend_Yield = float(num_span) if (num_span != '') else 0.0

            if 'ROCE' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                ROCE = float(num_span) if (num_span != '') else 0.0

            if 'ROE' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                ROE = float(num_span) if (num_span != '') else 0.0

            if 'Face Value' in name_span.text: 
                num_span = li.find('span',{'class':'number'})
                num_span = num_span.text.replace(',', '')
                Face_Value = float(num_span) if (num_span != '') else 0.0

                break

        print(f'MARKET CAPITILIZATION - {SCRIP[i]}: {market_cap} Cr')
        print(f'Current Price - {SCRIP[i]}: {Current_Price} Cr')
        print(f'High / Low - {SCRIP[i]}: {High} / {Low} Cr')
        print(f'Stock P/E - {SCRIP[i]}: {Stock_PE} Cr')
        print(f'Book Value - {SCRIP[i]}: {Book_Value} Cr')
        print(f'Dividend Yield - {SCRIP[i]}: {Dividend_Yield} %')
        print(f'ROCE - {SCRIP[i]}: {ROCE} Cr')
        print(f'ROE - {SCRIP[i]}: {ROE} %')
        print(f'Face Value - {SCRIP[i]}: {Face_Value}')
    
    except:
        print(f'EXCEPTION THROWN: UNABLE TO FETCH DATA FOR {SCRIP[i]}')
```
